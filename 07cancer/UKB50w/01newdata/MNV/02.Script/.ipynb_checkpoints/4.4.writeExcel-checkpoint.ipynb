```python
import xlrd
import xlwt
import numpy as np
```

```python
cancerList = ["01.Breast","02.Prostate","03.Melanoma","04.Colon","05.Lung","06.Cervical","07.Uterine","08.Bladder",
              "09.Rectal","10.Ovary","11.Kidney"]
```

```python
### MNVs AF , MNVs 包含的 SNVs
mnvFreqDict = {}
mnvSnvDict = {}
for cancer in cancerList:
    cancerName = cancer.split(".")[1]
    with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/"+cancer+"/sigGWA.5e-6.results.format.mnv","r") as fi:
        for line in fi:
            data = line.strip().split()
            datasetInfoList = data[8].split("|")
            mnvSnvDict[data[2]] = data[5].split(",")[:]
            for dataset in datasetInfoList:
                if dataset[:6] == "UKB50w":
                    AF = dataset.split("/")[1]
                    mnvFreqDict[data[2]] = "{:.4f}".format(eval(AF))
    fi.close()

### SNVs AF
snvFreqDict = {}
with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/01.Data/05.SigSNVsGenoData/sigSNVs.5e-6.afreq","r") as fi:
    fi.readline()
    for line in fi:
        data = line.strip().split()
        snvFreqDict[data[1]] = "{:.4f}".format(eval(data[4]))
fi.close()
    
### survival P
survPDict = {}
with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/all.11Cancers.survival.p.txt","r") as fi:
    fi.readline()
    for line in fi:
        data = line.strip().split()
        if data[1] not in survPDict:
            survPDict[data[1]] = {}
            survPDict[data[1]][data[0]] = "{:.4f}".format(eval(data[2]))
        else:
            survPDict[data[1]][data[0]] = "{:.4f}".format(eval(data[2]))
fi.close()

### annotation
annoDict = {}
for cancer in cancerList:
    cancerName = cancer.split(".")[1]
    with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/"+cancer+"/02.Anno.5e-6/en_GRCh37.75_gene.anno","r") as fi:
        for line in fi:
            data = line.strip().split("\t")
            if data[2] not in annoDict:
                annoDict[data[2]] = {}
                annoDict[data[2]][cancerName] = data[:]
            else:
                annoDict[data[2]][cancerName] = data[:]
    fi.close()
```

```python
### 分癌症写入excel表格
# 创建excel表格类型文件
book = xlwt.Workbook(encoding='utf-8', style_compression=0)

for cancer in cancerList:
    cancerName = cancer.split(".")[1]
    sheet = book.add_sheet(cancer, cell_overwrite_ok=True) # 创建不同表单

    cols = ['CHR','MNVs','POS','A1','A2','N','T','SE_T','P_noSPA','BETA','SE','P',"MNVs_AF","SNVs_AF","KM_P","Gene_Annotation"]
    for n in range(len(cols)):
        sheet.write(0, n, cols[n])
        
    with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/"+cancer+"/sigGWA.5e-6.results","r") as fi:
        lines = fi.readlines()
        for j in range(len(lines)):
            data = lines[j].strip().split("\t")
            SNVsList = mnvSnvDict[data[1]]
            myList = [annoDict[data[1]][cancerName][0] ,data[1], annoDict[data[1]][cancerName][1],
                      annoDict[data[1]][cancerName][3], annoDict[data[1]][cancerName][4], data[5], 
                      "{:.2f}".format(eval(data[7])), "{:.2f}".format(eval(data[8])), "{:.2e}".format(eval(data[9])),
                      "{:.4f}".format(eval(data[10])), "{:.4f}".format(eval(data[11])), "{:.2e}".format(eval(data[12])),
                      mnvFreqDict[data[1]], ",".join([snvFreqDict[SNVsList[0]],snvFreqDict[SNVsList[1]]]),
                      survPDict[data[1]][cancerName],annoDict[data[1]][cancerName][6]]
            for k in range(len(myList)):
                sheet.write(j + 1, k, myList[k])
    fi.close()
book.save('/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/sigGWA.5e-6.results.xls')
```

```python
### 所有癌症结果写入一页excel表格
# 创建excel表格类型文件
book = xlwt.Workbook(encoding='utf-8', style_compression=0)
sheet = book.add_sheet("allResults", cell_overwrite_ok=True)

a = 0
for cancer in cancerList:
    cancerName = cancer.split(".")[1]

    cols = ['Cancer Type','CHR','MNVID','POS','A1','A2','N','T','SE_T','P_noSPA','BETA','SE','P',"MNVs_AF","SNVs_AF","KM_P","Feature type","Gene"]
    for n in range(len(cols)):
        sheet.write(0, n, cols[n])
        
    with open("/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/"+cancer+"/sigGWA.5e-6.results","r") as fi:
        lines = fi.readlines()
        for j in range(len(lines)):
            data = lines[j].strip().split("\t")
            SNVsList = mnvSnvDict[data[1]]
            annoList = annoDict[data[1]][cancerName][6].split(" ")
            if len(annoList) == 1:
                featureType = "intergenic_variant"
                gene = "-"
            else:
                featureType = annoList[6]
                gene = annoList[1]
            myList = [cancer,annoDict[data[1]][cancerName][0] ,data[1], annoDict[data[1]][cancerName][1],
                      annoDict[data[1]][cancerName][3], annoDict[data[1]][cancerName][4], data[5], 
                      "{:.2f}".format(eval(data[7])), "{:.2f}".format(eval(data[8])), "{:.2e}".format(eval(data[9])),
                      "{:.4f}".format(eval(data[10])), "{:.4f}".format(eval(data[11])), "{:.2e}".format(eval(data[12])),
                      mnvFreqDict[data[1]], ",".join([snvFreqDict[SNVsList[0]],snvFreqDict[SNVsList[1]]]),
                      survPDict[data[1]][cancerName],featureType,gene]
            for k in range(len(myList)):
                sheet.write(j + a + 1, k, myList[k])
        a += len(lines)
    fi.close()
book.save('/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/03.Results/sigGWA.5e-6.allcancer.results.xls')
```
