```python
cancerList = ["01.Breast","02.Prostate","03.Melanoma","04.Colon","05.Lung","06.Cervical","07.Uterine","08.Bladder","09.Rectal","10.Ovary",
              "11.Kidney"]
singleSexCancerList = ["01.Breast","02.Prostate","06.Cervical","07.Uterine","10.Ovary"]

### 获取所有质控后白人样本
whitePeopleDict = {}
with open("/home/luohh/UKB50wOldData/02.PhenoData/ukb50.sex.height.popualtion.sample","r") as fi:
    fi.readline()
    for line in fi:
        data = line.strip().split()
        if data[3] in ["1","1001","1002","1003"]:
            whitePeopleDict[data[0]] = 1
fi.close()

dirString = "/home/luohh/MNVAnalysis/01.GWAS/02.Cancer/01.Data/02.PhenoData/"
for cancer in cancerList:
    #os.system("mv "+dirString+cancer+"/GWAS.last.covar "+dirString+cancer+"/GWAS.last.covar.bak")
    #os.system("mv "+dirString+cancer+"/GWAS.last.gcta.sample "+dirString+cancer+"/GWAS.last.gcta.sample.bak")
    #os.system("mv "+dirString+cancer+"/GWAS.last.pheno "+dirString+cancer+"/GWAS.last.pheno.bak")
    #os.system("mv "+dirString+cancer+"/GWAS.last.qcovar "+dirString+cancer+"/GWAS.last.qcovar.bak")
    #os.system("mv "+dirString+cancer+"/GWAS.last.sample "+dirString+cancer+"/GWAS.last.sample.bak")
    
    ### 质量性状（covar）
    fo = open(dirString+cancer+"/GWAS.last.covar","w")
    with open(dirString+cancer+"/GWAS.last.covar.bak","r") as fi:
        for line in fi:
            data = line.strip().split()
            if data[0] in whitePeopleDict:
                fo.write(line)
    fi.close()
    fo.close()
    
    ### 数量性状（qcovar）
    fo = open(dirString+cancer+"/GWAS.last.qcovar","w")
    with open(dirString+cancer+"/GWAS.last.qcovar.bak","r") as fi:
        for line in fi:
            data = line.strip().split()
            if data[0] in whitePeopleDict:
                if cancer in singleSexCancerList:
                    fo.write("\t".join(data[:24])+"\n")
                else:
                    fo.write("\t".join(data[:26])+"\n")
    fi.close()
    fo.close()
    
    ### 表型（pheno）
    fo = open(dirString+cancer+"/GWAS.last.pheno","w")
    with open(dirString+cancer+"/GWAS.last.pheno.bak","r") as fi:
        for line in fi:
            data = line.strip().split()
            if data[0] in whitePeopleDict:
                fo.write(line)
    fi.close()
    fo.close()
    
    ### 样本（gcta sample）
    fo = open(dirString+cancer+"/GWAS.last.gcta.sample","w")
    with open(dirString+cancer+"/GWAS.last.gcta.sample.bak","r") as fi:
        for line in fi:
            data = line.strip().split()
            if data[0] in whitePeopleDict:
                fo.write(line)
    fi.close()
    fo.close()
    
    ### 样本（sample）
    fo = open(dirString+cancer+"/GWAS.last.sample","w")
    with open(dirString+cancer+"/GWAS.last.sample.bak","r") as fi:
        for line in fi:
            data = line.strip().split()
            if data[0] in whitePeopleDict:
                fo.write(line)
    fi.close()
    fo.close()
```
